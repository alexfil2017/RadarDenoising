```python
import numpy as np
import scipy
import scipy.misc
import scipy.signal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = scipy.misc.imread("./data/test/Set12/08.png").astype(np.float32)
```

```python
plt.imshow(img, cmap='gray')
ker = 5*np.random.rand(4,4)
```

```python
img2 = scipy.signal.convolve2d(img, ker, mode='same', boundary='wrap')
img3 = scipy.signal.fftconvolve(img, ker, mode='same')
print(img2.shape)
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.imshow(img2, cmap='gray')
plt.subplot(132)
plt.imshow(img3, cmap='gray')
plt.subplot(133)
plt.imshow(img3-img2, cmap='gray')
```
